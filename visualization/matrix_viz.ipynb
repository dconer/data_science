### 1. imgshow for mapping a matrix to color

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Initialize columns and rows
m = 3
n = 5
M = np.random.randint(10,size=(m,n))
print(M)

# Mapping the matrix into color
plt.imshow(M)

# Write down the coefficients of the matrix into the color matrix
for i in range(m):
    for j in range(n):
        plt.text(j,i,str(M[i,j]),horizontalalignment='center',fontsize=28, verticalalignment='center')
        
# Add a colorbar
plt.colorbar()
plt.show()
```

### 2. imread to store a image in a matrix form

```python
from imageio import imread 

# To store a image in a matrix form in img
img = imread('https://upload.wikimedia.org/wikipedia/en/8/86/Einstein_tongue.jpg')

# Show image
plt.imshow(img)
```

```python
res = plt.hist(img.ravel(),bins=100)
```

### 3. heatmap diagram with seaborn

```python
from IPython import display 
display.set_matplotlib_formats('svg')
```

```python
import seaborn as sns
# Initialize variables
size = 10
hilb_mat = np.zeros((size,size))
# Create Hilbert matrix
for i in range(size):
    for j in range(size):
        hilb_mat[i,j] = 1/(i+j+1)
# Visualize matrix
sns.heatmap(hilb_mat,cmap="Blues", vmax=0.1)
plt.savefig('heatmap.pdf')
plt.show()
hilb_mat
```

```python
# Improving image's quality
from IPython import display 
display.set_matplotlib_formats('svg')

import random as r
import matplotlib.pyplot as plt

# Initialize variables
N = 10000
x = np.zeros(N)
y = np.zeros(N)
for i in range(N-1):
    k = r.randint(1,3)
    x[i+1] = 0.5*x[i]+k-1
    y[i+1] = 0.5*y[i]

    if k == 2:
        y[i+1]= y[i]+2
plt.style.use('fivethirtyeight')
plt.plot(x,y,marker='.',color='k',linestyle='none',markersize=1)
plt.show()
```
